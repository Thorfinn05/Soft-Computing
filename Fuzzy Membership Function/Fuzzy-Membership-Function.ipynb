```python
import numpy as np
import matplotlib.pyplot as plt
```

<h3>GAUSSIAN MEMBERSHIP FUNCTION</h3>

```python
def gaussian_mf(x, mean, sigma):
    return np.exp(-((x - mean) ** 2) / (2 * sigma * 2))

x = np.linspace(0, 10, 1000)

# m = float(input("Enter mean: "))
# s = float(input("Enter sigma: "))
y = gaussian_mf(x, 5.0, 2.0)

plt.plot(x, y, label=f'Gaussian MF (mean={5.0}, sigma={2.0})')
plt.xlabel('x values')
plt.ylabel('Membership Degree')
plt.title('Gaussian Membership Function')
plt.legend()
plt.grid()
plt.show()
```

<h3>TRIANGULAR MEMBERSHIP FUNCTION</h3>

```python
def triangular_membership(x, a, b, c):
    # if a <= x <= b:
    #     return (x-a)/(b-a)
    # elif b < x < c:
    #     return (c-x)/(c-b)
    # return 0.0
    return np.maximum(0, np.minimum((x-a)/(b-a), (c-x)/(c-b)))

x = np.linspace(0, 10, 1000)

# v_triangular = np.vectorize(triangular_membership)
# a = float(input("Enter a: "))
# b = float(input("Enter b: "))
# c = float(input("Enter c: "))
y = triangular_membership(x, 2.0, 4.0, 6.0)
plt.plot(x, y, label=f"Triangular MF ({2.0}, {4.0}, {6.0})")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Triangular Membership Function")
plt.grid()
plt.legend()
plt.show()
```

<h3>TRAPEZOIDAL MEMBERSHIP FUNCTION</h3>

```python
def trapezoidal_mf(x, a, b, c, d):
    if a <= x < b:
        return (x - a) / (b - a)
    elif b <= x <= c:
        return 1.0
    elif c < x <= d:
        return (d - x) / (d - c)
    return 0.0

x = np.linspace(0, 10, 1000)    
# a = float(input("Enter a: "))
# b = float(input("Enter b: "))
# c = float(input("Enter c: "))
# d = float(input("Enter d: "))

v_trap = np.vectorize(trapezoidal_mf)
y = v_trap(x, 2.0, 4.0, 6.0, 8)

plt.plot(x, y, label=f'Trapezoidal MF ({2.0}, {4.0}, {6.0}, {8.0})')
plt.title('Trapezoidal Membership Function')
plt.xlabel("x values")
plt.ylabel('Mmbership Degree')
plt.legend()
plt.grid()
plt.show()
```

<h3>GENERALISED BELL MEMBERSHIP FUNCTION</h3>

```python
def generalisd_bell_mf(x, a, b, c):
    return 1 / (1 + np.abs((x - c) / a) ** (2 * b))

x = np.linspace(0, 10, 1000)

# a = float(input("Enter width: "))
# b = float(input("Enter slope: "))
# c = float(input("Enter centre: "))
y = generalisd_bell_mf(x, 2.0, 4.0, 5.0)

plt.plot(x, y, label=f'width={2.0}\nslope={4.0}\ncentre={5.0}')
plt.xlabel('x values')
plt.ylabel('Membership Degree')
plt.title("Generalised Bell Mmbership Function")
plt.legend(); plt.grid()
plt.show() 
```

<h3>SIGMOID MEMBERSHIP FUNCTION</h3>

```python
def sigmoid_mf(x, a, c):
    return (1 / (1 + np.exp(-a * (x - c))))

x = np.linspace(-10, 10, 100)

# a = float(input('Enter slope: '))
# c = float(input('Enter centre: '))
y = sigmoid_mf(x, 1.0, 0.0)

plt.plot(x, y, label=f'Sigmoid MF\nSlope={1.0}, Centre={0.0}')
plt.title('Sigmoid Membership Function')
plt.xlabel('x values')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid()
plt.show()
```
