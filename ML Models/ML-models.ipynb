```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

X, y = load_iris(return_X_y = True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(y_pred)
acc = accuracy_score(y_test, y_pred)
print("Accuracy: ", round(acc, 4))

plt.bar(["Decision Tree"], [acc])
plt.ylabel("Accuracy")
plt.show()

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
plt.show()

plt.figure(figsize=(15, 10))
plot_tree(model,
          feature_names=load_iris().feature_names,
          class_names=load_iris().target_names,
        #   filled=True,
          rounded=True)
plt.title("Decision Tree")
plt.show()
```

```python
from sklearn.svm import SVC

model=SVC()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

acc = accuracy_score(y_test, y_pred)

print("Accuracy: ", round(acc, 4))

plt.bar(["SVM"], [acc])
plt.show()

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(max_iter=500)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print("Accuracy: ", round(acc, 2))

plt.bar(["MLP"], [acc])
plt.show()

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
plt.show()
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X, y = load_diabetes(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

score = r2_score(y_test, y_pred)
print("R2 score: ", round(score, 4))

# acc = accuracy_score(y_test, y_pred)
# print("Accuracy: ", round(acc, 4))

plt.bar(["Linear Regression"], [score])
plt.ylabel("R2 score")
plt.show()

# sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
# plt.show()
print(load_diabetes().feature_names)
import numpy as np
X_bmi = X[:, np.newaxis, 2]
X_train, X_test, y_train, y_test = train_test_split(X_bmi, y, test_size=0.3)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='black', label='Actual Data')
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Regression Line')
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix

X, y = load_iris(return_X_y=True)

kmeans = KMeans(n_clusters=3)
y_pred = kmeans.fit_predict(X)

sns.heatmap(confusion_matrix(y, y_pred), annot=True)
plt.title("Confusion Matrix")
plt.show()

X_vis = X[:, :2]
kmeans_vis = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans_vis.fit_predict(X_vis)

plt.figure(figsize=(10, 6))
plt.scatter(X_vis[y_kmeans == 0, 0], X_vis[y_kmeans == 0, 1], s=70, c='salmon', label='Cluster 1')
plt.scatter(X_vis[y_kmeans == 1, 0], X_vis[y_kmeans == 1, 1], s=70, c='lightblue', label='Cluster 2')
plt.scatter(X_vis[y_kmeans == 2, 0], X_vis[y_kmeans == 2, 1], s=70, c='lightgreen', label='Cluster 3')

# Plot the centroids
plt.scatter(kmeans_vis.cluster_centers_[:, 0], kmeans_vis.cluster_centers_[:, 1], 
            s=250, marker='*', c='black', label='Centroids')

plt.title('K-Means Clustering (Iris Sepal Features)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```
